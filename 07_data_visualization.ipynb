```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn liberary
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare')
```

```python
df.plot(kind='scatter', x='age', y='fare', color= 'red')
```

```python
# change the colour using hex code
df.plot(kind='scatter', x='age', y='fare', color= '#F67B0E')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Fare')
```

```python
# Seaborn Plots

# Bar plot using seaborn on df
sns.barplot(data= df,x='survived', y='fare' )
```

```python
sns.scatterplot(data= df,y='fare', x='age', hue='sex' )
```
