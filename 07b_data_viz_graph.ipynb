### 1-Creating line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
# data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
#         'Sales': [100, 120, 110, 130, 140]}
df = sns.load_dataset('titanic')

# Create a line plot
df.plot(x='survived', y='fare', kind='line')

# Show the plot
plt.show()
```

#### Interpretation of Line Graph
Line graph between survived & Fare shows that some survived persons pay more fare as compare to not survived persons.
