## Matplotlib Chart

### Mat_Graph_01

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Bar Chart: Survival Rate by Class
plt.figure(figsize=(10, 6))
df.groupby(['pclass', 'sex'])['alive'].value_counts(normalize=True).unstack().plot(kind='bar')
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

first class female survived more as compared to not survived.\
first class male survived less as compared to not survived.\
second class female survived more as compared to not survived.\
first class male survived less as compared to not survived.\
third class female survived and not survived are equal.\
The third class male survived less as compared to not survived.

### Mat_Graph_02

```python
# 2. Histogram: Age Distribution
plt.figure(figsize=(10, 6))
df['age'].plot(kind='hist', bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

children of age 0 to 5 are around 40 in numbers.\
children of age 5 to 10 are around 15 in numbers.\
children of age 10 to 15 are around 10 in numbers.\
There are majority of young people on ship aged between 20 to 30.

### Mat_Graph_03

```python
# 3. Pie Chart: Embarkation Towns
plt.figure(figsize=(10, 6))
df['embark_town'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Embarkation Towns')
plt.show()
```

Pie chart shows that more than 70% passenger embark from southampton followed by cherbourg and Queenstown.

### Mat_Graph_04

```python
# 4. Scatter Plot: Fare vs Age
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['fare'])
plt.title('Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

the scatter plot shows that passeneger of all age group mostly paid fare upto 100 pounds.

### Mat_Graph_05

```python
# 5. Bar Chart: Survival Rate by Sex
plt.figure(figsize=(10, 6))
df['alive_num'] = df['alive'].map({'yes': 1, 'no': 0})  # Map 'yes' to 1 and 'no' to 0
df.groupby(['sex'])['alive_num'].mean().plot(kind='bar')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

Survival rate of female is much higher than male.

### Mat_Graph_06

```python
# 6. Box Plot: Fare by Class
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='fare', data=df)
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

the box plot shows that first class has values from 0 to 180.\
second class has values from 0 to 50 pounds.\
third class has values from 0 to 20.

### Mat_Graph_07

```python
# 7. Heatmap: Correlation Matrix
import numpy as np
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

there is weak negative relationship of -0.34 between survived & pclass mean as pclass increase by one unit the survival decrease by 0.34 unit.\
there is very weak negative relationship of -0.077 between survived & age mean as age increase by one unit the survival decrease by 0.077 unit.\
there is very weak negative relationship of -0.035 between survived & sibsp mean as sibsp increase by one unit the survival decrease by 0.035 unit.\
there is very weak positive relationship of 0.082 between survived & parch mean as parch increase by one unit the survival incraese by 0.082 unit.\
there is weak positive relationship of 0.26 between survived & fare mean as fare increase by one unit the survival incraese by 0.26 unit.\
there is strong positive relationship of 1 between survived & alive_num mean as alive_num increase by one unit the survival incraese by 1 unit.



### Mat_Graph_08

```python
# 8. Bar Chart: Survival Rate by Pclass and Sex
plt.figure(figsize=(10, 6))
df['alive_num'] = df['alive'].map({'yes': 1, 'no': 0})  # Map 'yes' to 1 and 'no' to 0
df.groupby(['pclass', 'sex'])['alive_num'].mean().unstack().plot(kind='bar')
plt.title('Survival Rate by Pclass and Sex')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

In all classes the survival rate of females are high as compared to male.
