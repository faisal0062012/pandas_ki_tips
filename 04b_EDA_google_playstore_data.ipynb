```python
# Convert the whole size column into bytes

# Let's define a function

def convert_size(size):
    if isinstance(size, str): 
        if 'k' in size:
            return float(size.replace('k', "")) * 1024
        elif 'M' in size:
            return float(size.replace('M', "")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
# Let's apply this function
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

```python
# rename the column
df.rename(columns={'Size': "sizes_in_bytes"}, inplace=True)
```

```python
df.head()
```

```python
df['sizes_in_Mb'] = df['sizes_in_bytes'].apply(lambda x: x/(1024*1024))
```

```python
df['sizes_in_kb'] = df['sizes_in_bytes'].apply(lambda x: x/(1024))
```

- The only problem is + sign in suffix of numericals which we will remove 

# Installs column
1. Remove '+' Sign
2. remove ','
3. Convert column into an integer

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+','') if '+' in str(x) else x)
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',','') if ',' in str(x) else x)
```

```python
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

```python
df['Installs'].value_counts()
```

# Price Column

```python
df['Price'].value_counts()
```

```python
# Let's see how many values have '$' sign
df['Price'].loc[df['Price'].str.contains('\$')].value_counts().sum()
```

```python
# This code count the number of values in 'Price' column which contain 0 but does not contain $ sign
df['Price'].loc[(df['Price'].str.contains('0')) & (~df['Price'].str.contains('\$'))].value_counts().sum()
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$','') if '$' in str(x) else x)
```

```python
df['Price'].value_counts()
```

```python
# Now we can convert this to a numeric value
df['Price'] = df['Price'].apply(lambda x: float(x))
```

```python
df.describe()
```

```python
# using f string print min, max and average price of the app
print(f'Min price is {df['Price'].min()}')
print(f'Max price is {df['Price'].max()}')
print(f'Average price is {df["Price"].mean()}')
```

# Missing Values

```python
# find missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# find missing values percentage in the data
missing_values_percentage = round((df.isnull().sum() / len(df))*100, 2).sort_values(ascending=False)
print(missing_values_percentage)
```

```python
# Find total number of missing values
df.isnull().sum().sum()
```

```python
# plot missing values
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull().astype(int), cmap='YlGnBu', cbar=False)
```

```python
# make figure size
plt.figure(figsize=(12, 6))
# plot the null values by their percentage in each column
missing_percentage = df.isnull().sum()/len(df) * 100
missing_percentage.plot(kind='bar')
# add the labels
plt.xlabel('columns')
plt.ylabel('percentage')
plt.title('Percentage of missing values in each column')
# sns.heatmap(df.isnull().astype(int), cmap='YlGnBu', cbar=False)
```
